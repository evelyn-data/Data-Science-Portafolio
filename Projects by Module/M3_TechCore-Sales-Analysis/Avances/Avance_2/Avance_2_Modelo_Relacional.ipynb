# Modelo Relacional TechCore

## Objetivo

El objetivo es crear un Modelo Relacional con pandas.
Este cuaderno transforma un archivo csv desnormalizado (tabla plana) en un archivo Excel listo para cargar a PowerBI.

### Importación de Librerías y Carga de Datos


```python
# Importar llibrerías
import pandas as pd
import numpy as np
```

```python
# Cargar archivio csv desde Google Drive

file_id = '1dAoCeJtwqOCaWGQxcn0uOSmgbV0hZrHL'
url = f'https://drive.google.com/uc?id={file_id}'

df = pd.read_csv(url, sep=",")
```

```python
# Analisis breve de las columnas
df.info()
```

```python
# Mostrar las primeras filas del DataFrame
df.head()
```

### Crear Tablas Normalizadas

En esta estapa definimos las entidades y los atributos para crear las tablas que conformarán la estructura del modelo, a su vez se definen las claves foráneas y primarias.





### Tabla Ciudades

```python
Ciudades =  df['CiudadSucursal2'].unique()
Ciudades = pd.DataFrame(Ciudades, columns=['Ciudad'])
Ciudades['CiudadID'] = Ciudades.index + 1
Ciudades = Ciudades[['CiudadID', 'Ciudad']]
print("---Tabla Ciudades---")
Ciudades
```

### Tabla de Sucursales

```python
# Crear la tabla Sucursales
Sucursales = df[['SucursalNombre', 'CiudadSucursal2']].drop_duplicates().reset_index(drop=True)
#Clave primaria SucursalID
Sucursales['SucursalID'] = Sucursales.index + 1
# Agreagamos la llave foranea CiudadID
Sucursales = Sucursales.merge(Ciudades, left_on='CiudadSucursal2', right_on='Ciudad', how='left')
Sucursales = Sucursales[['SucursalID', 'SucursalNombre', 'CiudadID']]
print("---Tabla Sucursales---")
Sucursales
```

### Tabla Clientes

```python
Clientes = df[['ClienteNombre', 'EmailCliente', 'EdadCliente', 'GeneroCliente', 'TelefonoCliente', 'DireccionCliente']].drop_duplicates().reset_index(drop=True)
#Clave primaria ClienteID
Clientes['ClienteID'] = Clientes.index + 1
Clientes = Clientes[['ClienteID', 'ClienteNombre', 'EmailCliente', 'EdadCliente', 'GeneroCliente', 'TelefonoCliente', 'DireccionCliente']]
print("---Tabla Clientes---")
Clientes.head()
```

```python
Clientes['ClienteNombre'].value_counts()
```

### Tabla Vendedores

```python
Vendedores = df['VendedorNombre'].unique()
Vendedores = pd.DataFrame(Vendedores, columns=['VendedorNombre'])
# Clave primaria VendedorID
Vendedores['VendedorID'] = Vendedores.index + 1
Vendedores = Vendedores[['VendedorID', 'VendedorNombre']]
Vendedores.head()
```

### Tabla Método Pago

```python
MetodosPago = df['MetodoPago'].unique()
MetodosPago = pd.DataFrame(MetodosPago, columns=['MetodoPago'])
MetodosPago['MetodoPagoID'] = MetodosPago.index + 1
MetodosPago = MetodosPago[['MetodoPagoID', 'MetodoPago']]
MetodosPago
```

### Tabla Facturas


```python
Facturas = df[['VentaID', 'Year', 'Month', 'HoraVenta', 'FechaVenta', 'SucursalNombre',
               'ClienteNombre', 'EmailCliente', 'VendedorNombre', 'MetodoPago',
               'DescuentoVenta','TotalVenta']].drop_duplicates().reset_index(drop=True)

Facturas =  Facturas.merge(Sucursales, left_on='SucursalNombre', right_on='SucursalNombre', how='left')
Facturas = Facturas.merge(Vendedores, left_on='VendedorNombre', right_on='VendedorNombre', how='left')
Facturas = Facturas.merge(Clientes, left_on=['ClienteNombre', 'EmailCliente'], right_on=['ClienteNombre', 'EmailCliente'], how='left')
Facturas = Facturas.merge(MetodosPago, left_on='MetodoPago', right_on ='MetodoPago', how='left')


Facturas = Facturas[['VentaID', 'Year', 'Month', 'HoraVenta', 'FechaVenta', 'SucursalID',
                     'VendedorID', 'ClienteID', 'MetodoPagoID', 'DescuentoVenta','TotalVenta']]

Facturas
```

### Tabla Productos

```python
# Crear la tabla Productos
Productos = pd.DataFrame(columns=['NombreProducto', 'MarcaProducto', 'Precio'])

# Iteramos sobre los productos en el DataFrame original y los agregamos a la tabla Productos
for x in range (1,4):
    append_producto = pd.DataFrame(columns= ['NombreProducto', 'MarcaProducto', 'Precio'])
    append_producto['NombreProducto'] = df[f'NombreProducto{x}']
    append_producto['MarcaProducto'] = df[f'MarcaProducto{x}']
    append_producto['Precio'] = df[f'PrecioUnitarioProducto{x}']

    Productos = pd.concat([Productos, append_producto], ignore_index=True)

Productos = Productos.dropna().drop_duplicates().reset_index(drop=True)
# Clave primaria ProductoID
Productos['ProductoID'] = Productos.index + 1
Productos = Productos[['ProductoID', 'NombreProducto', 'MarcaProducto', 'Precio']]
print("---Tabla Productos---")
Productos.head()
```

### Tabla de Detalles

```python

DetalleFacturas = pd.DataFrame(columns=['VentaID', 'ProductoID', 'Cantidad', 'Subtotal'])

for i in range(1, 4):
    detalle = df[['VentaID', f'NombreProducto{i}',f'MarcaProducto{i}', f'CantidadProducto{i}', f'SubtotalProducto{i}']]
    # Clave foranea ProductoID
    detalle = detalle.merge(Productos, left_on=[f'NombreProducto{i}', f'MarcaProducto{i}'],
                        right_on=['NombreProducto', 'MarcaProducto'], how='left')
    detalle = detalle[['VentaID', 'ProductoID', f'CantidadProducto{i}', f'SubtotalProducto{i}']]
    detalle = detalle.rename(columns={f'CantidadProducto{i}': 'Cantidad', f'SubtotalProducto{i}': 'Subtotal'})
    
    DetalleFacturas = pd.concat([DetalleFacturas, detalle], ignore_index=True)


DetalleFacturas = DetalleFacturas.dropna().reset_index(drop=True)


# DetalleFacturas.info()

# Ajustar tipos de datos
DetalleFacturas['Cantidad'] = DetalleFacturas['Cantidad'].astype(int)
DetalleFacturas['Subtotal'] = DetalleFacturas['Subtotal'].astype(float)
DetalleFacturas['VentaID'] = DetalleFacturas['VentaID'].astype(int)
DetalleFacturas['ProductoID'] = DetalleFacturas['ProductoID'].astype(int)

DetalleFacturas.sort_values(by=['VentaID'], inplace=True)
DetalleFacturas.reset_index(drop=True, inplace=True)

# Clave primaria DetalleID
DetalleFacturas['DetalleID'] = DetalleFacturas.index + 1


DetalleFacturas = DetalleFacturas[['DetalleID', 'VentaID', 'ProductoID', 'Cantidad', 'Subtotal']]

print("---Tabla DetalleFacturas---")
DetalleFacturas.head()

```

## Análisis Exploratorio


### Ventas Por Marca

```python
# Ventas por marca

pd.options.display.float_format = '{:,.2f}'.format

Ventas_por_marca = DetalleFacturas.merge(Productos, on= 'ProductoID')

Ventas_por_marca = Ventas_por_marca.groupby('MarcaProducto')['Subtotal'].sum().reset_index()

Ventas_por_marca = Ventas_por_marca.sort_values(by='Subtotal', ascending=False)

print('---Ventas por marca---')
Ventas_por_marca
```

Lenovo es la marca que más vende.

### Productos Más Vendidos

```python
#10 Productos más vendidos

Ventas_por_producto = DetalleFacturas.merge(Productos, on= 'ProductoID')
Ventas_por_producto = Ventas_por_producto.groupby('NombreProducto')['Subtotal'].sum().reset_index()
Ventas_por_producto = Ventas_por_producto.sort_values(by='Subtotal', ascending=False)
print('---10 Productos más vendidos---')
Ventas_por_producto.head(10)
```

### Distribución de ventas por ciudad

```python
### Distribución de ventas por ciudad

df_analisis =  Facturas.merge(Sucursales, on='SucursalID')

df_analisis = df_analisis.merge(Ciudades, on='CiudadID')

ventas_por_ciudad = df_analisis.groupby('Ciudad')['TotalVenta'].sum().reset_index()

ventas_por_ciudad

total_global = ventas_por_ciudad['TotalVenta'].sum()

ventas_por_ciudad['% Participación'] = (ventas_por_ciudad['TotalVenta'] / total_global) * 100

ventas_por_ciudad = ventas_por_ciudad.sort_values(by='TotalVenta', ascending=False).reset_index(drop=True)

pd.options.display.float_format = '{:,.2f}'.format
print("Distribución de Ventas por Ciudad:")
print(ventas_por_ciudad)

```

**Conclusión**

El modelo relacional permite realizar un análisis exploratorio consistente, evitando duplicados o errores de ambiguedad.

## Exportar Modelo a Archivo Excel

```python
dataframes = [
    Ciudades,
    Sucursales,
    Clientes,
    Vendedores,
    MetodosPago,
    Facturas,
    Productos,
    DetalleFacturas
]


nombrar_hojas = [
    'Ciudades',
    'Sucursales',
    'Clientes',
    'Vendedores',
    'MetodosPago',
    'Facturas',
    'Productos',
    'DetalleFacturas'
]


for df, nombre_hoja in zip(dataframes, nombrar_hojas):
    df.fillna(' ', inplace=True)

ruta_archivo = 'ModeloVentas.xlsx'

with pd.ExcelWriter(ruta_archivo, engine='xlsxwriter') as writer:
    for df, nombre_hoja in zip(dataframes, nombrar_hojas):
        df.to_excel(writer, sheet_name=nombre_hoja, index=False)
```
